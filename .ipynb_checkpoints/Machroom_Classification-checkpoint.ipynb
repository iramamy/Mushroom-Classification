## Encode all column

```python
from sklearn.preprocessing import LabelEncoder
```

```python
def encode_column(col_name):
    le = LabelEncoder()
    le.fit(col_name)

    return le.transform(col_name)
```

```python
col_names = data.columns
for col_name in col_names:
    data[col_name] = encode_column(data[col_name])
```
